# 4a

```python
x, y, z, a, r, t = sp.symbols("x y z a r t")
```

```python
lim1 = 0
lim1 = sp.Eq(x, 0)
lim1 = lim1.subs({x: r*sp.cos(t)})
lim1 = sp.solve(lim1, r)
lim1 = 0
lim1
sp.Eq(r, lim1)
```

```python
lim2 = sp.sqrt(a**2-y**2)
lim2 = sp.Eq(x, lim2)
lim2 = lim2.subs({x: r*sp.cos(t), y: r*sp.sin(t)})
lim2 = sp.solve(lim2, r)
lim2 = a
lim2
sp.Eq(r, lim2)
```

Converting dy to dr gives r=0 and r=a. This corresponds to da where r=0 and r=a. This shows that the area on the xy plane is within 0 to r in the positive x and positive y quadrant. This means it is a quarter circular shape on the xy plane. Since the dz remains dz during conversion the z=+-a indicates that this is a cylinder not a sphere. This means the object is a quarter cylinder with radius a in the positive xy quadrant, that stretches from z=-a to z=a.

```python
lim3 = 0
lim3 = sp.Eq(y, lim3)
lim3 = lim3.subs({y: r*sp.sin(t)})
lim3 = sp.solve(lim3, t)
lim3 = 0
lim3
sp.Eq(t, lim3)
```

```python
lim4 = sp.pi/2
lim4
sp.Eq(t, lim4)
```

```python
lim5 = -a
lim5
```

```python
lim6 = a
lim6
```

```python
f = r
f = sp.integrate(f, (r, lim1, lim2))
f = sp.integrate(f, (t, lim3, lim4))
f = sp.integrate(f, (z, lim5, lim6))
f
```

```python
x, y, z, a, r, t, p, ph = sp.symbols("x y z a r t p ph")
```

```python
f = 1
f = sp.integrate(f, (p, 1, 3))
f = sp.integrate(f, (t, 0, 2*sp.pi))
f = sp.integrate(f, (ph, 0, sp.pi))
f
```

```python
f = p**2*sp.sin(ph)
f = sp.integrate(f, (p, 1, 3))
f = sp.integrate(f, (t, 0, 2*sp.pi))
f = sp.integrate(f, (ph, 0, sp.pi))
f
```
