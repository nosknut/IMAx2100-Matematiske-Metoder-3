```python
rotate = -(45*(sp.pi/180))
gap = 0.95
lx, ly = sp.symbols("x, y")
spl.plot3d_parametric_surface(
    (x, y, 2*x, (fi, 0, sp.pi), (th, 0, 2*sp.pi*gap)),
    (x, y, 2*y, (fi, 0, sp.pi), (th, 0, 2*sp.pi*gap)),
    nb_of_points_u=10,
    nb_of_points_y=10
)

spl.plot3d_parametric_surface(
    (lx, ly, 2*lx, (lx, -10, 10), (ly, -10, 10)),
    (lx, ly, 2*ly, (lx, -10, 10), (ly, -10, 10)),
    nb_of_points_u=10,
    nb_of_points_y=10
)

spl.plot3d_parametric_surface(
    (x, y, z, (fi, 0, sp.pi), (th, 0+rotate, 2*sp.pi*gap+rotate)),
    nb_of_points_u=10,
    nb_of_points_y=10
)

spl.plot3d_parametric_surface(
    (x, y, z, (fi, 0, sp.pi), (th, 0+rotate, 2*sp.pi*gap+rotate)),
    (lx, ly, 2*lx, (lx, -10, 10), (ly, -10, 10)),
    nb_of_points_u=10,
    nb_of_points_y=10
)

spl.plot3d_parametric_surface(
    (x, y, z, (fi, 0, sp.pi), (th, 0+rotate, 2*sp.pi*gap+rotate)),
    (lx, ly, 2*ly, (lx, -10, 10), (ly, -10, 10)),
    nb_of_points_u=10,
    nb_of_points_y=10
)
```

```python
x, y = sp.symbols("x, y")

dx = sp.Derivative(x)/x
dy = sp.Derivative(y)/y
nabla = sp.Matrix([dx, dy])
r = sp.Matrix([2*x*y, 2*y*x**2])
```

```python
r
```

```python
nabla.dot(r)
```

```python
nabla.dot(r).doit()
```

```python
x, y, z = sp.symbols("x, y, z")

dx = sp.Derivative(x)/x
dy = sp.Derivative(y)/y
dz = sp.Derivative(z)/z
nabla = sp.Matrix([dx, dy, dz])
r = sp.Matrix([2*x*y, 2*y*x**2, z**4*x**2+y])
```

```python
nabla
```

```python
r
```

```python
nabla.cross(r)
```

```python
nabla.cross(r).doit()
```
