```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = range(1, 6)
```

```python
def f(starting, interest, payments):
    ys = [float(starting)]
    for x in xs:
        value = ys[x-1]
        ys.append(value + value*interest/12 - payments)
    return ys
```
